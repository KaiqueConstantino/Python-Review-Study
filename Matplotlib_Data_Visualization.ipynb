```python
import matplotlib.pyplot as plt 
```

```python
# Creating an line graph
x = [1, 2, 3, 4]
y = [10, 20 , 25, 30]

# Creating the graph
plt.plot(x, y)
plt.title("Line Graph")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.show()
```

```python
# Bar graph
category = ["A", "B", "C", "D"]
values = [10, 20, 30, 40]


# Creating the plot 
plt.bar(category, values)
plt.title("Bar graph")
plt.show()
```

```python
# Pizza graph
category = ["A", "B", "C"]
values = [10, 20, 30]

# Creating the plot 
plt.pie(values, labels=category)
plt.title("Pizza Graph")
plt.show()
```

```python
"""    
Exerccio 1: matplotlib
Crie um grfico de barras que mostre as vendas de 4 produtos diferentes.
Crie um grfico de pizza que mostre a participao percentual de cada produto nas vendas totais.

"""

product = ["Macbook", "Iphone", "Samsung book", "Samsumg Galaxy"]
sales = [200, 1000, 300, 1100]

# Creating the plot 
plt.bar(product, sales, color="skyblue")
plt.title("Total sales Percentage")
plt.xlabel("Products")
plt.ylabel("Sales")
plt.show()


# Creating the pizza Graph

plt.pie(sales, labels=sales, autopct="%1.1f%%")
plt.title("Products percentage")
plt.show()
```
