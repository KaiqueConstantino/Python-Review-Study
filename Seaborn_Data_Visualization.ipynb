```python
import seaborn as sns
import pandas as pd
```

```python
# Creating a dataframe
df = pd.DataFrame({
    "Altura": [1.60, 1.65, 1.70, 1.75, 1.80, 1.85, 1.90, 1.95, 2.00],
    "Peso": [60, 65, 70, 75, 80, 85, 90, 95, 100]
})

# Dispersal Plotting
sns.scatterplot(x = "Altura", y = "Peso", data = df)
```

```python
# Creating a barplot 

df2 = pd.DataFrame({
    "Produto": ["Smartphone", "Tablet", "Laptop"],
    "Vendas": [200, 500, 1000]
})

sns.barplot(x = "Produto", y = "Vendas", data = df2)
```
